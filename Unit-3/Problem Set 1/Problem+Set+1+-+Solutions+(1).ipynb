## Problem: Largest Prime Factor

Write a C++ program that finds the largest prime factor of any positive integer greater than 1. You must use the algorithm described in class where you successively divide by the next smallest factor until only 1 remains.

```c++
a = 5491; // 17*17*19
int currentLargest = 1;
int currentTest = 2;

while(currentLargest <= a){
    if(a % currentTest == 0){ // currentLargest is a prime factor
        currentLargest = currentTest;
        a /= currentTest;
    }else{
        currentTest++;
    }
}
cout << currentLargest;
```

## Problem: Fibonocci Sequence
### Part 1

The fibonocci sequence is defined as:  
$$\left\{\begin{matrix}

\\ f_n = f_{n-1} + f_{n-2}
\end{matrix}\right.$$

The first 5 terms are as follows: 1, 1, 2, 3, 5.  

Write a code chunk to print the first 10 Fibonocci Numbers.

```c++
int fn_2;
int fn_1 = 1;
int n = 10;
int fn = 1;

cout << 1;

for(int i = 1; i < n; i++){
    fn_2 = fn_1;
    fn_1 = fn;
    cout << ", " << fn;
    fn = fn_1 + fn_2;
}
```

## Problem: Palindromic Numbers

### Part 1
A palindromic number reads the same both ways. For example 9, 232, and 7007 are all palindromic numbers.

Write a code chunk that checks whether any positive integer $m$ is palindromic and outputs the result as follows:
* "Is a palindrome"
* "Is not a palindrome"

```c++
int m = 111222111;

int mRemaining = m;
int reverse = 0;

while(mRemaining > 0){
    reverse = reverse * 10 + mRemaining % 10;
    mRemaining /= 10;
}

cout << m << ", " << reverse << endl;

if(m == reverse)
    cout << "Is a palindrome";
else
    cout << "Is not a palindrome";
```

### Part 2

Find the largest palindromic number made from the product of two 4-digit numbers.


```c++
int largest = 0;
int current, reverse;

for(int i = 1000; i <= 9999; i++){
    for(int j = 1000; j <= 9999; j++){
        // now i*j is a product of two four digit numbers
        current = i * j;
        reverse = 0;
        while(current > 0){
            reverse = reverse * 10 + current % 10;
            current /= 10;
        }
        if(reverse == i * j && reverse > largest){
            // we have found the new biggest palindrome
            largest = reverse;
        }
    }
}
cout << largest;
```
