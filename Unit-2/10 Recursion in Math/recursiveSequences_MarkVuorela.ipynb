## Example

### Question 1a.

$
\left\{
   \begin{matrix}
      a_1 = 4 \\ 
      a_n = n + a_{n-1}+6
   \end{matrix}
\right.
$

First 6 terms: 4, 12, 21, 31, 42, 54

```c++
int question1a(int n){
    if(n == 1)
        return 4;
    else
        return n + 6 + question1a(n-1);
}
```

```c++
cout << "The first " << TERMS << " terms are: ";

for(int i = 1; i <= TERMS; i++){
    cout << question1a(i) << " ";
}
```
