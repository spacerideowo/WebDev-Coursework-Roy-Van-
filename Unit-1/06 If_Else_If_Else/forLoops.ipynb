
# For Loops (same diagram as while loops)

<img src="https://static.javatpoint.com/core/images/java-while-loop.png" alt="While Loop" />

Let's start with an example to see what it does:

````c++
// Count to 10
for(int i = 1; i <= 10; i++)
    cout << i << endl;
````

Let's breakdown some of the above lines of code. What does the following do?

`int i = 1;`

This sets thi initial value of the counter. In this case i. One thing to notice is that i only exists in the for loop. For example, the following code will generate an error.

```
for(int i = 1; i <= 10; i++)
    cout << i << endl;

cout << i << endl;
```

This is because `i` only exists within the loop.

The next part `i <= 10;` gives the exit condition of the loop. Once this condition evaluates to `false` the loop stops.

The last part `i++` is the update part. This part will be run **at the end** of each loop.

Some of last lesson's exercises are better suited for a for loop. Let's revisit each:

## Why have both while and for loops?

For loops are a tidy way to repeat a task when we know how many times we will do it:
Examples:
- Repeat something 10 times.
- Move from the first element of a list to the last element of a list.
- Move through each letter in a word or sentence.

While loops are better for more complicated situations where we will repeat something an unknown number of times.
- Getting input from a user
- Problems involving division or multiplication
- Waiting for a change such as a number becoming negative

## Consider the following code:

```
int x = 10;

while(x >= 0)
   cout << x << endl;
```

What's wrong with the previous code?

**Answer**  
Since x is never changed `x >= 10` will always be true. This leads to what is called an **infinite loop** because the loop will not end on its own.

# More Common C++ Shortcuts

So far we've seen `x++`, `++x`, `--x` and `x--`.
C++ offers other convient shortcuts that you can use. 

Check the following code block for examples. All of these examples do two things in one shortcut. 
1. Perform a mathematical operation
2. Store the result

For example, the following two lines of code do the same thing:
```
x += 10;
x = x + 10;
```

````c++
int x;

cout << "Increasing x by 6.\n";x = 10;

cout << "Before: x = " << x << endl;

x += 6;
cout << "After:  x = " << x << endl;
````

````c++
int x;

cout << "Decreasing x by 10.\n";
x = 10;
cout << "Before: x = " << x << endl;\
x -= 10;
cout << "After:  x = " << x << endl;
````

````c++
int x;

cout << "Dividing x by 2.\n";
x = 10;
cout << "Before: x = " << x << endl;\
x /= 2;
cout << "After:  x = " << x << endl;
````

````c++
int x;

cout << "Multiplying x by 2.\n";
x = 10;
cout << "Before: x = " << x << endl;\
x *= 2;
cout << "After:  x = " << x << endl;
````
