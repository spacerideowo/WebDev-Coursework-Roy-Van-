# Introduction to C++: Comments and Integers

## Include the input/output stream

```c++
#include <iostream>

using namespace std;
```

## Comments in C++
### Single Line comments

```c++
// Single line comments look like this

cout << 23 << endl;
// cout << 46 << endl;
```

### Multi-line comments

```c++
/*
    Sometimes we want to have large comment chunks
    We can do this with block comments.
*/

cout << 1 << endl;
/*
cout << 2 << endl;
cout << 3 << endl;
*/
cout << 4 << endl;
```

## Integers
### Integer Variables
Variables in C++ (and other programming languages) are similar to those in math. We can use variables to store data and then access the data later. Here is how we can create integer variables:

```c++
// We can declare a single integer variable
int x;
// Or multiple variables.
int y, z;
```

```c++
// C++ will not give a warning if we try to access a new variable:
cout << x << endl;
```

```c++
// We can initialize variables (set their values) them as follows:
x = 2;
y = 3; 
z = 5;
```

```c++
// Let's print out these values
cout << "x = " << x << endl;
cout << "y = " << y << endl;
cout << "z = " << z << endl;
```

### Variable Operations with Integers

```c++
// Integer addition with literals (actual numbers)
cout << "7 + 11 = "<< 7 + 11 << endl;
// Integer addition with variables (stored numbers)
cout << "x + y = "<< x + y << endl;
// And a combination of both
cout << "7 + x = " << 7 + x << endl;
```

```c++
// Integer subtraction:
cout << "x - y = " << x - y << endl;
```

```c++
// Integer division
cout << "12 / 6 = " << 12 / 6 << endl;
cout << "x / y = " << x / y << endl;
```

```c++
// Integer multiplication
cout << "12 * 6 = " << 12 * 6 << endl;
cout << "x * y = " << x * y << endl;
```

```c++
// Modulo division (finds the remainder)
cout << "12 % 2 = " << 12 % 2 << endl;
cout << "12 % 5 = " << 12 % 5 << endl;
cout << "x % y = " << x % y << endl;
```

## Comparison

```c++
// Comparing two integers
cout << "12 == 12: "<< (12 == 12) << endl;
cout << "12 == 13: "<< (12 == 13) << endl;
```

```c++
cout << "true: " << true << endl;
cout << "false: " << false << endl;
```

```c++
// is 13 odd?
cout << "13 % 2 == 1: " << (13 % 2 == 1) << endl;
```

```c++
// is 101 even?
cout << "101 % 2 == 0: " << (101 % 2 == 0) << endl;
```

### Let's look at <, <=, >, >=, !=

```c++
cout << "x = " << x << endl;
cout << "y = " << y << endl;
```

```c++
// less than
cout << "x < y: " << (x < y) << endl;
```

```c++
// greater than
cout << "x > y: " << (x > y) << endl;
```

```c++
// less than or equal to
cout << "x + 1 <= y: " << (x + 1 <= y) << endl;
```

```c++
// greater than or equal to
cout << "x >= y: " << (x >= y) << endl;
```

```c++
// not equal to
cout << "x != y: " << (x != y) << endl;
```

## Characters

```c++
char a, b, c;
a = 'x';
b = 'Y';
c = '5';
```

```c++
cout << a << ", " << b << ", " << c << endl;
```

```c++
// characters are stored as integers, so we can see their values (note that x is an integer)
x = a;
cout << x << endl;
```

```c++
// we can add to characters like we would integers
a = a + 1;
cout << a << endl;
```

```c++
// if we go too far, we don't know what we will get (if anything...)
// a = a + 100; // this breaks the code!
cout << a << endl;
```

```c++
x = 'a';
cout << x << endl;
x = 'A';
cout << x << endl;
```

### Converting from lowercase to uppercase (using math)
Since 'a' is 97 and 'A' is 65 we can convert from lowercase to uppercase by subtracting 97 - 65 = 32

```c++
// x is an integer
x = 'a' - 'A';
```

```c++
// b is a character
b = 's' - x;
cout << b << endl;
```
